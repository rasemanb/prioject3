```python
import pandas as pd

import json

from sqlalchemy import create_engine

import numpy as np

from config import user, passw, host, port, name
```

```python
gdp = pd.read_csv("data/gdp.csv", skiprows=4)
gni = pd.read_csv("data/gni.csv", skiprows=4)
coord = pd.read_csv("data/coord.csv")
debt = pd.read_csv("data/debt.csv")
```

```python
coord = coord[['name', 'country', 'latitude', 'longitude']]
coord.rename(columns={"name":"country_name",
                     "country":"country_code",
                     "latitude":"lat",
                     "longitude":"lon"}, inplace=True)
coord.head()
```

```python
coord_final = coord.loc[coord["country_name"].isin(["United States",
                                         "China", 
                                         "Japan",
                                         "Germany", 
                                         "India", 
                                         "United Kingdom", 
                                         "France", 
                                         "Italy", 
                                         "Canada", 
                                         "South Korea"])]
coord_final.loc[119, 'country_name'] = "Korea, Rep."
```

```python
coord_final.dtypes
```

```python
debt.drop(columns= ['place', 
            'pop2023', 
            'area', 
            'country', 
            'cca2', 
            'ccn3', 
            'region', 
            'subregion', 
            'landAreaKm', 
            'density', 
            'densityMi', 
            'Rank', 
            'rank'], inplace=True)

debt.rename(columns={
    "growthRate":"growth_rate",
    "cca3":"country_code",
    "debt2gdpRatio":"debt_to_gdp_ratio"
}, inplace=True)

debt = debt[['country_code', 'growth_rate', 'debt_to_gdp_ratio']]

debt
```

```python
debt_final = debt.loc[debt["country_code"].isin(["USA",
                                         "CHN", 
                                         "JPN",
                                         "DEU", 
                                         "IND", 
                                         "GBR", 
                                         "FRA", 
                                         "ITA", 
                                         "CAN", 
                                         "KOR"])]

debt_final.sort_values("debt_to_gdp_ratio", ascending=False, inplace=True)

country = ['Japan', 
           'Italy', 
           'United States', 
           'France', 
           'Canada', 
           'United Kingdom', 
           'India',
           'Germany', 
           'China', 
           'Korea, Rep.']
debt_final['country_name'] = country

debt_final = debt_final[['country_name', 
                         'country_code', 
                         'growth_rate', 
                         'debt_to_gdp_ratio']]

debt_final
```

```python
debt_final.reset_index(inplace=True)
```

```python
debt_final.drop("index", axis=1, inplace=True)
```

```python
protocol = 'postgresql'
rds_connection_string = f'{protocol}://{user}:{passw}@{host}:{port}/{name}'
engine = create_engine(rds_connection_string)
```

```python
engine.table_names()
```

```python
gdp_final.to_sql(name='gdp', con=engine, if_exists='append', index=False)
gni_final.to_sql(name='gni', con=engine, if_exists='append', index=False)
coord_final.to_sql(name='coord', con=engine, if_exists='append', index=False)
debt_final.to_sql(name='debt', con=engine, if_exists='append', index=False)
```
