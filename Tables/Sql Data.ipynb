```python
import pandas as pd 
import json
from sqlalchemy import create_engine
import numpy as np
from config import pw
```

```python
csv_file = "./Resources/GDP.csv"
gdp_df = pd.read_csv(csv_file)
gdp_df
```

```python
data_df = gdp_df[['Country Name', 'Series Name', '2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]', '2021 [YR2021]']].copy()
data_df
```

```python
data_df.replace('..', np.nan, inplace=True)

drop_df = data_df.dropna(how='any')
drop_df
```

```python
rename_df = drop_df.rename(columns = {"2012 [YR2012]":"2012",
                         "2013 [YR2013]":"2013", 
                         "2014 [YR2014]":"2014", 
                         "2015 [YR2015]":"2015", 
                         "2016 [YR2016]":"2016",
                         "2017 [YR2017]":"2017",
                         "2018 [YR2018]":"2018",
                         "2019 [YR2019]":"2019",
                         "2020 [YR2020]":"2020",
                         "2021 [YR2021]":"2021"})
rename_df
```

```python
numbers = rename_df.loc[drop_df['Country Name'].isin(['United States', 'China', 'Japan', 'Germany', 'India', 'United Kingdom', 'France', 'Italy', 'Canada', 'Korea, Rep.'])]
numbers
```

```python
numbers.reset_index(inplace=True)
numbers
```

```python
numbers.drop("index", axis=1, inplace=True)
numbers
```

```python

numbers.loc[17, 'Country Name'] = "South Korea"
numbers
```

```python
numbers.loc[7, 'Country Name'] = "South Korea"
numbers
```

```python
data = numbers[numbers["Series Name"] == "GDP (current US$)"]
data
```

```python
pd.to_numeric(data ['2021']).sort_values()
```

```python
gdp_new = data[["Country Name", "Series Name"]].copy(deep=True)
for year in ["2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021"]:
    gdp_new[year] = data[year].apply(pd.to_numeric)
```

```python
gdp = gdp_new.sort_values(by = ['2012'], ascending=(False))
gdp
```

```python
data2 = numbers[numbers["Series Name"] == "GNI (current US$)"]
data2
```

```python
pd.to_numeric(data2 ['2021']).sort_values()
```

```python
gni_new = data2[["Country Name", "Series Name"]].copy(deep=True)
for year in ["2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021"]:
    gni_new[year] = data2[year].apply(pd.to_numeric)
```

```python
gni = gni_new.sort_values(by = ['2021'], ascending=(False))
gni
```
